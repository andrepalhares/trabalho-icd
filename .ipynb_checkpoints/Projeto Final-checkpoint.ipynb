# Teste de hipótese

#### Há mais pessoas com doutorado ganhando mais de 50k do que pessoas com bacharelado?

```python
filtro_bach = df['Education'] == ' Bachelors'
df[filtro_bach]['Salary'].mean()

filtro_doc = df['Education'] == ' Doctorate'
df[filtro_doc]['Salary'].mean()

t_obs = df[filtro_doc]['Salary'].mean() - df[filtro_bach]['Salary'].mean()

N = 1000
difer = np.zeros(N)
for i in range(N):
    np.random.shuffle(filtro_bach.values)
    diff = df[filtro_doc]['Salary'].mean() - df[filtro_bach]['Salary'].mean()
    difer[i] = diff
        
plt.hist(difer, bins=50, edgecolor='k')
plt.xlabel('Diferença na Permutação')
plt.ylabel('Pr(diff)')
plt.show()
```

# Treinando modelo de regressão linear

Primeiro vamos separar o dataset em dois grupos: um contendo as colunas referentes aos atributos pessoais das pessoas e outro com a coluna de salário. Depois, vamos dividir cada um dos grupos em dois grupos: um de teste e outro de treinamento. No final, vamos avaliar a regressão linear obtida.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def converte_bool_salary(x):
    return 0 if x == True else 1

def converte_bool_sex(x):
    return 1 if str(x) == ' Male' else 0

def converte_pais(pais):
    if pais == ' United-States':
        return 6
    elif pais == ' Outlying-US(Guam-USVI-etc)':
        return 5
    elif pais == ' Canada':
        return 4
    elif (pais == ' England' or pais == ' Italy' or pais == ' France' or pais == ' Portugal' or 
    pais == ' Germany' or pais == ' Greece' or pais == ' Poland' or pais == ' Ireland'
    or pais == ' Holand-Netherlands' or pais == ' Scotland' or pais == ' Yugoslavia' or pais == ' Hungary'):
        return 3
    elif (pais == ' Vietnam' or pais == ' Taiwan' or pais == ' Hong' or pais == ' Japan'
    or pais == ' Thailand' or pais == ' India' or pais == ' China' or pais == ' Philippines'
    or pais == ' Iran' or pais == ' Cambodia' or pais == ' Laos'):
        return 2
    elif (pais == ' Mexico' or pais == ' Dominican-Republic' or pais == ' Trinadad&Tobago'
    or pais == ' El-Salvador' or pais == ' Columbia' or pais == ' Puerto-Rico' or pais == ' Honduras'
    or pais == ' Guatemala' or pais == ' Peru' or pais == ' Ecuador' or pais == ' Haiti'
    or pais == ' Nicaragua' or pais == ' Jamaica' or pais == ' Cuba'):
        return 1
    else:
        return 0

def converte_workclass(work):
    if work == ' Private':
        return 8
    elif work == ' Self-emp-not-inc':
        return 7
    elif work == ' ?':
        return 6
    elif work == ' Local-gov':
        return 5
    elif work == ' State-gov':
        return 4
    elif work == ' Federal-gov':
        return 3
    elif work == ' Self-emp-inc':
        return 2
    elif work == ' Without-pay':
        return 1
    elif work == ' Never-worked':
        return 0
    
def converte_etnia(x):
    if x == ' White':
        return 4
    elif x == ' Black':
        return 3
    elif x == ' Asian-Pac-Islander':
        return 2
    elif x == 'Amer-Indian-Eskimo':
        return 1
    else:
        return 0
```

```python
df_regressao_linear = pd.read_csv('salaries.csv')
df_regressao_linear['Salary'] = df_regressao_linear['Salary'].apply(converte_bool_salary)
df_regressao_linear['Sex'] = df_regressao_linear['Sex'].apply(converte_bool_sex)
df_regressao_linear['NativeCountry'] = df_regressao_linear['NativeCountry'].apply(converte_pais)
df_regressao_linear['Workclass'] = df_regressao_linear['Workclass'].apply(converte_workclass)
df_regressao_linear['Ethnicity'] = df_regressao_linear['Ethnicity'].apply(converte_etnia)
df_regressao_linear.head()
```

```python
# Dvidindo dataset em dois grupos

atributos_pessoais = df_regressao_linear[['Age', 'EducationNum', 'HoursPerWeek', 'CapitalGain',
                                          'CapitalLoss', 'Sex', 'NativeCountry', 'Workclass', 
                                          'Ethnicity']]
salarios = df_regressao_linear['Salary']
```

```python
# Dividindo cada um dos grupos em dois outros grupos, um de teste e outro de treinamento
x_train, x_test, y_train, y_test = train_test_split(atributos_pessoais, salarios, test_size=0.3, random_state=101)
```

```python
# Criando modelo de regressão linear
regressao_linear = LinearRegression()

regressao_linear.fit(x_train,y_train)
```

# Treinando modelo de regressão linear


Semelhante ao treino anterior, vamos separar o dataset em dois grupos: um contendo as colunas referentes aos atributos pessoais dos entrevistados e outro com a coluna de salário. Depois, vamos dividir cada um dos grupos em dois grupos: um de teste e outro de treinamento.
No final, vamos analisar se é possível prever se uma pessoa tem o salário superior a 50 mil dólares anuais ou não.

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Modelando os dados
atributos_pessoais = df_regressao_linear[['Age', 'EducationNum', 'HoursPerWeek', 'Ethnicity',
                                          'CapitalLoss', 'Sex', 'NativeCountry', 'Workclass', 
                                          'CapitalGain']]
salarios = df_regressao_linear['Salary']
```

```python
# Dividindo cada um dos grupos em dois outros grupos, um de teste e outro de treinamento
x_train, x_test, y_train, y_test = train_test_split(atributos_pessoais, salarios, test_size=0.3, random_state=101)
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
```

```python
# Agora vamos prever o quão bom é nosso teste
from sklearn.metrics import classification_report, confusion_matrix

confusion_matrix(y_test, predictions)
```
